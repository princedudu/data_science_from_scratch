```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(20)

k1 = (86 - 111)/(4.2 - 0) # slope of the first line
b1 = 111 # intercept of the first line

k2 = (117 - 165)/(4.2 - 0) # slope of the second line
b2 = 165 # intercept of the second line


fig, ax = plt.subplots()
ax.plot(x, k1 * x + b1)
ax.plot(x, k2 * x + b2)
# ax.annotate('Something', xy=(x[0], y[0]))
ax.annotate('Intersection point', xy = (9.860, 52.304), textcoords='data')

plt.legend(['First line', 'Second line'], loc = 'upper right')
plt.show()
```

```python
# intersection point

# k1 * x_0 + b1 = k2 * x_0 + b2
x_0 = (b2 - b1)/(k1 - k2)
y_0 = k1 * x_0 + b1
print(x_0, y_0)
```
